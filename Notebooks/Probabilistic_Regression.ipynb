```python
import numpy as np
import matplotlib.pyplot as plt

# generate data
x = np.random.rand(30)
e = np.random.randn(30) * 0.1
a = 0.3
y = (a*x) + e

# approximation
a_estimate = np.sum(x * y) / np.sum(x**2)
x_range = np.linspace(0, 1, 30)
y_regr = x_range * a_estimate
plt.scatter(x, y)
plt.plot(x_range, y_regr, color="orange")
```
