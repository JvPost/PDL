```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 2
x = np.random.rand(N)
y = 0.3 * x
noise = np.random.randn(N) / 10
x+=noise
plt.scatter(x, y)
```

```python
x_range = np.linspace(0, 1, N)
a = np.sum(y * x) / np.sum(x**2)
y_hat = a * x_range
plt.plot(x_range, y_hat)
plt.scatter(x, y)
a
```
