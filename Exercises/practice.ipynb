```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import auc
```

```python
d_x = 0.001
X = np.arange(-3, 3, 0.01)
def f(x):
    return np.e**(-(x**2))


def run(X, d_x, f, plot = True, normalize = True):
    Y = np.array([f(x) for x in X])
    if normalize:
        Y = 1/auc(X, Y) * Y
    if plot:
        plt.plot(X, Y)   
    return Y 
```

```python
Y = run(X, d_x, f)
auc(X, Y)
```

```python
auc(X, Y)
```

```python
X = np.arange(0, 5, d_x)

def f(x):
    return np.exp(-x)

Y = run(X, d_x, f)
```

```python
X = np.arange(1, 20, d_x)

def f(x):
    return 1/x

Y = run(X, d_x, f)
```

```python
auc(X, Y)
```
